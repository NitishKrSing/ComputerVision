```python
f = misc.face(gray = True)
plt.imshow(f, cmap = plt.cm.gray)
```

```python
f.shape
```

Contrast

```python
plt.imshow(f, cmap = plt.cm.gray, vmin=30, vmax = 200)
plt.show()
```

Mirror image of the above image

```python

mirror_face = np.fliplr(f)
plt.imshow(mirror_face, cmap = plt.cm.gray)
plt.show()
```

```python
mirror_face = np.flip(f)
plt.imshow(mirror_face, cmap = plt.cm.gray, vmin = 30, vmax =200)
plt.show()
```
